```python
import nltk
from nltk.translate import AlignedSent, IBMModel1
from nltk.corpus import comtrans

# Load a small corpus from NLTK
nltk.download('comtrans')

# Create aligned sentences (parallel corpus)
aligned_sents = comtrans.aligned_sents()

# Train a Statistical Machine Translation model (IBM Model 1)
ibm_model = IBMModel1(aligned_sents, 1)  # Number of iterations

# Translate a sentence manually using the translation model
source_sentence = input("enter sentence : ")
source_words = source_sentence.split()

# Translate each word in the sentence using IBM Model 1
translated_words = []
for word in source_words:
    # Get the translation probabilities for the word
    translation_probs = ibm_model.translation_table[word]

    # Select the best translation based on the highest probability
    best_translation = max(translation_probs, key=translation_probs.get)

    translated_words.append(best_translation)

print("Original Sentence:", source_sentence)
print("Translated Sentence:", " ".join(translated_words))
```

```python
# 1. Install Gradio (Run this in a Colab cell)
!pip install gradio -q

import gradio as gr
import nltk
from nltk.translate import IBMModel1
from nltk.corpus import comtrans

# 2. Download the data
nltk.download('comtrans')

# 3. Load English-French aligned sentences
# Using a specific file prevents mixing multiple languages together
print("Loading data and training model... please wait.")
als = comtrans.aligned_sents('alignment-en-fr.txt')

# 4. Train IBM Model 1
# We use 10 iterations to get slightly better accuracy than just 1
ibm_model = IBMModel1(als, 10)

def translate_logic(input_sentence):
    if not input_sentence.strip():
        return "Please enter a sentence."

    # Process words (lowercase to match training data)
    source_words = input_sentence.lower().split()
    translated_words = []

    for word in source_words:
        # Safety check: if word exists in the model's vocabulary
        if word in ibm_model.translation_table:
            probs = ibm_model.translation_table[word]
            # Get the best translation (highest probability)
            best_trans = max(probs, key=probs.get)

            # If the model returns None, keep the original word
            translated_words.append(best_trans if best_trans else word)
        else:
            # Handle unknown words
            translated_words.append(f"[{word}]")

    return " ".join(translated_words)

# 5. Create the Gradio Interface
app = gr.Interface(
    fn=translate_logic,
    inputs=gr.Textbox(lines=2, placeholder="Enter English sentence here...", label="English Input"),
    outputs=gr.Textbox(label="French Translation (IBM Model 1)"),
    title="Statistical Machine Translation",
    description="This uses NLTK's IBM Model 1. Note: It translates word-for-word based on probability and may not follow French grammar."
)

# 6. Launch with a public link
app.launch(share=True)
```
